The objective of this project is to perform customer segmentation using the 'Credit Card Dataset for Clustering' from Kaggle. The dataset contains the usage behavior of approximately 9,000 active credit card holders over a six-month period. By applying clustering algorithms, the goal is to identify distinct customer segments based on their purchasing habits and credit card usage.

This analysis will aid in defining a marketing strategy by clustering customers based on features such as PURCHASES and CREDIT_LIMIT. Two approaches will be used for clustering: Hierarchical Clustering and K-means Clustering. The performance of the clusters will be evaluated, and the best number of clusters will be determined using the Elbow Method.

Key questions to address include:

1. How can we identify groups of customers with similar behavior?
2. What are the characteristics of the identified clusters?
3. How can the results of this analysis help in forming a data-driven marketing strategy?

```python
# importing key libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering, KMeans
import scipy.cluster.hierarchy as shc # to compute hierarchical clustering and generate dendrogram plots 
```

```python
# Loading the dataset
df = pd.read_csv('Credit_card_dataset.csv')
```

Data preprocessing

```python
# Dropping the CUST_ID column
df = df.drop(columns=['CUST_ID'])
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
# fill missing value with mean)
df.fillna(df.mean(), inplace=True)
```

```python
df.isnull().sum()
```

```python
# OUTLIER DETECTION
# Calculating Q1 (25th percentile) and Q3 (75th percentile)
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
```

```python
# Calculating IQR (Interquartile Range)
IQR = Q3 - Q1
```

```python
# Detecting outliers
outliers = df[((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]
print(f"Outliers detected:\n{outliers}")
```

```python
# Handling outliers(using capping)

# Calculating lower and upper bounds for each column
lower_bound = df.quantile(0.05)
upper_bound = df.quantile(0.95)

# Capping the outliers in each column
df_capped = df.clip(lower=lower_bound, upper=upper_bound, axis=0)
```

```python
# Standardizing the data before clustering
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df[['PURCHASES', 'CREDIT_LIMIT']])
```

```python
# Plotting the Dendrogram for Hierarchical Clustering

plt.figure(figsize=(10, 7))
plt.title("Dendrograms")
dend = shc.dendrogram(shc.linkage(df_scaled, method='ward'))
plt.show()
```

A dendrogram shows how clusters are merged and helps one determine the number of clusters; 3 for my case.

```python
# Applying Agglomerative Clustering
hc = AgglomerativeClustering(n_clusters=3, linkage='ward')
df['Cluster_HC'] = hc.fit_predict(df_scaled)
```

```python
# Plotting the clusters formed by Hierarchical Clustering
plt.figure(figsize=(10,6))
sns.scatterplot(x='PURCHASES', y='CREDIT_LIMIT', hue='Cluster_HC', data=df, palette='Set1')
plt.title('Hierarchical Clustering of Purchases vs Credit Limit')
plt.show()
```

visualizing the clusters formed by plotting PURCHASES against CREDIT_LIMIT with color indicating the cluster assignment.
Clusters Represent Different Customer Segments; Each cluster reflecting a different segment of the customer base, characterized by their credit card usage behavior.

# interpretation:
# cluster 0: 
Customers in this cluster may have low to moderate purchases and balances. They might represent budget-conscious customers who utilize their credit cards sparingly.
The marketing strategy for these would include targeting these customers with promotions to encourage more frequent use of their credit cards or suggest products that align with their spending patterns.
# cluster 1:
This cluster might include customers who have high credit limits and make large purchases. They are potentially higher-value customers who use their credit cards actively.
The marketing strategy here would entail engaging these customers with premium offers, higher credit limits, or exclusive access to products/services to maintain loyalty.
# cluster 2:
Customers here may show a mix of high cash advances and payments, indicating they are either managing debt or utilizing their cards for larger expenses. This group may also include riskier customers with higher balances.
For these one should develop tailored financial management tools, debt relief programs, or educational content on responsible credit card use to assist them.

```python
# Applying K-means
kmeans = KMeans(n_clusters=3, random_state=0)
df['Cluster_KMeans'] = kmeans.fit_predict(df_scaled)
```

```python
# Plotting the K-means clusters

plt.figure(figsize=(10,6))
sns.scatterplot(x='PURCHASES', y='CREDIT_LIMIT', hue='Cluster_KMeans', data=df, palette='Set2')
plt.title('K-means Clustering of Purchases vs Credit Limit')
plt.show()
```

```python
# Using the Elbow method to find the best K
sse = [] 
# SSE is used to find the optimal number of clusters by looking at the "elbow" in the graph
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(df_scaled)
    sse.append(kmeans.inertia_)

# Plotting the Elbow graph
plt.figure(figsize=(10,6))
plt.plot(range(1, 11), sse)
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

our K value remains as 3, therefore it is the optimum one as confirmed by the elbow method above.
